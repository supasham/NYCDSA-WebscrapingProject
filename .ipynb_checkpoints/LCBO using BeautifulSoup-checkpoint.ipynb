```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
url = "https://www.lcbo.com/webapp/wcs/stores/servlet/en/lcbo/wine-14/wine-14/red-wine-14001?pageView=grid&orderBy=4&fromPage=catalogEntryList&beginIndex=6084"
s = requests.session()
response = s.get(url, headers=headers)
```

```python
text = BeautifulSoup(response.text, 'html.parser')
```

```python
print(text.prettify())
```

```python
temp = 'spaced-character'
num_bottles = text.find('span', attrs={'class': temp}).string
num_bottles = int(num_bottles)
num_bottles
```

```python
url_list = []
url = 'https://www.lcbo.com/webapp/wcs/stores/servlet/en/sdp/red-wine-14001?pageView=grid&orderBy=4&fromPage=catalogEntryList&beginIndex='
for i in range(0, num_bottles, 12):
    url_list.append(url+str(i))
print(url_list[-5:])
```

```python
temp = 'row product'
bottles = text.find_all('div', attrs={'class': 'row product'})
print(len(bottles))
```

```python
bottle = bottles[0]

# Find name
name = bottle.find('div', attrs={'class': 'productChart'}).find('a').get_text()

# Find price
price = bottle.find('div', attrs={'class': 'productChart'}).find('span', attrs={'class': 'price'}).get_text()
price = float(price.strip()[1:].replace(',', ''))

# Find container type
cont = bottle.find('div', attrs={'class': 'other_details'}).find('span', attrs={'class': 'package-type'}).get_text()

# Find container size
size = bottle.find('div', attrs={'class': 'other_details'}).find('span').get_text()
size = re.findall('\d*\ ',size)[0].strip()

# Find region
region = bottle.find('div', attrs={'class': 'region hidden-sm'}).find('span').get_text().strip()

# Find critic
try:
    critic = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'critic'}).get_text().strip()
except:
    # critic not found
    critic = ''

# Find critic score
try:
    critic_score = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'score'}).get_text().strip()
    critic_score = re.findall('\d*',critic_score)[0].strip()
except:
    # critic score not found
    critic_score = 0
```

```python
# Scrape single page
import csv
with open('lcbo.csv', 'w', newline='') as csvfile:
    bottle_writer = csv.writer(csvfile)
    bottle_writer.writerow(['name','region','size-mL','container','price','critic','critic_score']) 
    for bottle in bottles:
        dic = {}
        # Find name
        name = bottle.find('div', attrs={'class': 'productChart'}).find('a').get_text()
        # Find price
        price = bottle.find('div', attrs={'class': 'productChart'}).find('span', attrs={'class': 'price'}).get_text()
        price = float(price.strip()[1:].replace(',', ''))
        # Find container type
        cont = bottle.find('div', attrs={'class': 'other_details'}).find('span', attrs={'class': 'package-type'}).get_text()
        # Find container size
        size = bottle.find('div', attrs={'class': 'other_details'}).find('span').get_text()
        size = re.findall('\d*\ ',size)[0].strip()
        # Find region
        region = bottle.find('div', attrs={'class': 'region hidden-sm'}).find('span').get_text().strip()
        # Find critic
        try:
            critic = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'critic'}).get_text().strip()
        except:
            # critic not found
            critic = 'none'
        # Find critic score
        try:
            critic_score = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'score'}).get_text().strip()
            critic_score = re.findall('\d*',critic_score)[0].strip()
        except:
            # critic score not found
            critic_score = 0
            
        dic['name'] = name
        dic['region'] = region
        dic['size'] = size
        dic['cont'] = cont
        dic['price'] = price
        dic['critic'] = critic
        dic['critic_score'] = critic_score
        
        bottle_writer.writerow(dic.values())         
```

```python
# Apply to all pages
import time
import random
import csv

def scrape_single_page(bottles, csvwriter):
    for bottle in bottles:
        dic = {}
        # Find name
        name = bottle.find('div', attrs={'class': 'productChart'}).find('a').get_text()
        # Find price
        price = bottle.find('div', attrs={'class': 'productChart'}).find('span', attrs={'class': 'price'}).get_text()
        price = float(price.strip()[1:].replace(',', ''))
        # Find container type
        cont = bottle.find('div', attrs={'class': 'other_details'}).find('span', attrs={'class': 'package-type'}).get_text()
        # Find container size
        size = bottle.find('div', attrs={'class': 'other_details'}).find('span').get_text()
        size = re.findall('\d*\ ',size)[0].strip()
        # Find region
        region = bottle.find('div', attrs={'class': 'region hidden-sm'}).find('span').get_text().strip()
        # Find critic
        try:
            critic = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'critic'}).get_text().strip()
        except:
            # critic not found
            critic = 'none'
        # Find critic score
        try:
            critic_score = bottle.find('div', attrs={'class': 'product_score'}).find('div', attrs={'class': 'score'}).get_text().strip()
            critic_score = re.findall('\d*',critic_score)[0].strip()
        except:
            # critic score not found
            critic_score = 0
            
        dic['name'] = name
        dic['region'] = region
        dic['size'] = size
        dic['cont'] = cont
        dic['price'] = price
        dic['critic'] = critic
        dic['critic_score'] = critic_score
        
        csvwriter.writerow(dic.values())  

        
        
# Webscrape LCBO for red wines      
s = requests.session()
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
with open('lcbo_redwine.csv', 'w', encoding='utf-8', newline='') as csvfile:
    bottle_writer = csv.writer(csvfile)
    bottle_writer.writerow(['name','region','size-mL','container','price','critic','critic_score']) 
    for index, url in enumerate(url_list):
        response = s.get(url, headers=headers)
        text = BeautifulSoup(response.text, 'html.parser')
        bottles = text.find_all('div', attrs={'class': 'row product'})
        scrape_single_page(bottles, bottle_writer)
        # Random sleep to avoid getting banned from the server
        time.sleep(random.randint(1,3))
        print('Finished page: ' + str(index+1))
s.close()        
```
